```python
import joblib
import pandas as pd

# Load test data
test_data = pd.read_excel('../Kaggle/input/widsdatathon2025/TEST/TEST_QUANTITATIVE_METADATA.xlsx')  # Replace with your test data file path

test_data
```

```python
X_test = test_data.drop(columns=['participant_id'])
```

```python
# Load the ADHD model
model_adhd = joblib.load('Models/model_adhd.pkl')

# Load the Sex model
model_sex = joblib.load('Models/model_sex.pkl')
```

```python
X_test.fillna(0, inplace=True)
```

```python
# Predict ADHD outcome
adhd_preds = model_adhd.predict(X_test)

# Predict Sex
sex_preds = model_sex.predict(X_test)
```

```python
len(adhd_preds)==len(sex_preds)
```

```python
# Create a DataFrame for the submission
submission = pd.DataFrame({
    'participant_id': test_data['participant_id'],
    'sex': sex_preds,
    'adhd': adhd_preds
})

# Save the DataFrame to a CSV file
submission.to_csv('submissions.csv', index=False)
```
