```python
import pandas as pd 
```

```python
quantiative_train_data = pd.read_excel('../Kaggle/input/widsdatathon2025/TRAIN_NEW/TRAIN_QUANTITATIVE_METADATA_new.xlsx')
quantiative_train_data.head()
```

```python
connctome_matrics = pd.read_csv('../Kaggle/input/widsdatathon2025/TRAIN_NEW/TRAIN_FUNCTIONAL_CONNECTOME_MATRICES_new_36P_Pearson.csv')
connctome_matrics.head()
```

```python
training_solutions = pd.read_excel('../Kaggle/input/widsdatathon2025/TRAIN_NEW/TRAINING_SOLUTIONS.xlsx')
training_solutions.head()
```

```python
# merged_df = pd.merge(quantiative_train_data, connctome_matrics,  on='participant_id', how='inner')  
# merged_df
merged_df = quantiative_train_data
```

```python
merged_df = pd.merge(merged_df, training_solutions,  on='participant_id', how='inner')  
merged_df
```

```python
merged_df = merged_df.dropna()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
ADHD = merged_df['ADHD_Outcome']
ADHD
```

```python
Sex = merged_df['Sex_F']
Sex
```

```python
X = merged_df.drop(columns=['ADHD_Outcome','Sex_F','participant_id'])
X
```

```python
X_train, X_test, ADHD_train, ADHD_test = train_test_split(X, ADHD, test_size=0.2, random_state=42)
```

```python
model_sex = LogisticRegression(max_iter=10000)
model_sex.fit(X_train, ADHD_train)
```

```python
X_train, X_test, Sex_train, Sex_test = train_test_split(X, Sex, test_size=0.2, random_state=42)
```

```python
model_adhd = LogisticRegression(max_iter=10000)
model_adhd.fit(X_train, Sex_train)
```

```python
import joblib

# Save the ADHD model
joblib.dump(model_adhd, 'Models/model_adhd.pkl')

# Save the Sex model
joblib.dump(model_sex, 'Models/model_sex.pkl')
```
